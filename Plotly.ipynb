```python
import plotly.graph_objs as go
import pandas as pd
from plotly.offline import iplot
```

```python
trace2=go.Scatter(x=df.world_rank,y=df.teaching,mode="markers+lines",
                  name="teaching",marker=dict(color='rgba(29,135,3,0.9)'),text=df.university_name)
```

```python
# Preparing the Data
df2014=my_data[my_data.year==2014].iloc[:100,:]
df2015=my_data[my_data.year==2015].iloc[:100,:]
df2016=my_data[my_data.year==2016].iloc[:100,:]
```

```python
track1=go.Scatter(x=df2014.world_rank,y=df2014.citations,
                  mode="lines",name="2014",marker=dict(color='rgba(219,15,15,1)'),text=df2014.university_name)
track2=go.Scatter(x=df2015.world_rank,y=df2015.citations,
                 mode="lines",name="2015",marker=dict(color='rgba(0,173,168,1)'),
                 text=df2015.university_name)

track3=go.Scatter(x=df2016.world_rank,y=df2016.citations,
                 mode="lines",name="2016",marker=dict(color='rgba(60,120,7,1)'),
                 text=df2016.university_name)
```

```python
data=[track1,track2,track3]
```

```python
layout=dict(title="World's University Rank Comparison with citations in various years",
            xaxis=dict(title="World's Rank",ticklen=6,zeroline=False),
           yaxis=dict(title="Citations",ticklen=5))
```

```python
fig2=dict(data=data,layout=layout)
```

```python
iplot(fig2)
```

```python
student_staff2011=my_data.student_staff_ratio[my_data.year==2011]
student_staff2012=my_data.student_staff_ratio[my_data.year==2012]
```
